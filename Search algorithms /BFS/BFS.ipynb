# Breadth first search (BFS) Pseudocode

##

# Strategy :
### Layer by layer strategy : expand all nodes at depth d before moving to depth d+1
   

# Data Structure : 
### uses Queue (FIFO) to manage the frontier (open list)

# Complexity : 
   ### branching factor (b) depth (d)
   ## Time : O(b^d) -> exponential
   ## Space : O(bd) -> keeps all nodes in memory

# Properties :
  ## Complete -> find solutions if exists
  ## Optimal  -> for uniform step costs
